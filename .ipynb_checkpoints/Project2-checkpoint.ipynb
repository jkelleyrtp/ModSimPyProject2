# Agent Based Modeling for Oncolyctic Virotherapy Drug Delivery
    - Oncolyctic Virus: Newcastle Virus, Reovirus, Parvovirus
### Drug Delivery Methods:
    "Pill Scenario" - complete random dispersal 
    "Injection Scenario" - Single point
    "Capsule Scenario" - cluster 
The goal of this project is to find the ideal amount and delivery method of oncolyctic cells to successfully kill the most tumor cells without affecting the overall homeostasis of the body. We want to minimize the negative side effects. We want to cure cancer. 
We want to make a grid of tissue (brain, heart, etc..).
